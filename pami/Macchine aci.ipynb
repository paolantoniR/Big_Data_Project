```python
import pandas as pd
```

```python
# Correctly load the file uploaded by the user for transformation
file_path = 'YOUR FILE PATH TO/macchine_aci.xlsx'

# Read the Excel file
aci_data = pd.read_excel(file_path)

# Desired regions to keep and renaming mappings
desired_regions = [
    'Lombardia', 'Emilia-Romagna', 'Toscana', 'Umbria', 'Lazio',
    'Campania', 'Piemonte', 'Liguria', 'Veneto', 'Marche', 'Abruzzo',
    'Puglia', 'Friuli-Venezia Giulia'
]

rename_mapping = {
    'LOMBARDIA': 'Lombardia',
    'EMILIA ROMAGNA': 'Emilia-Romagna',
    'TOSCANA': 'Toscana',
    'UMBRIA': 'Umbria',
    'LAZIO': 'Lazio',
    'CAMPANIA': 'Campania',
    'PIEMONTE': 'Piemonte',
    'LIGURIA': 'Liguria',
    'VENETO': 'Veneto',
    'MARCHE': 'Marche',
    'ABRUZZO': 'Abruzzo',
    'PUGLIA': 'Puglia',
    'FRIULI VENEZIA GIULIA': 'Friuli-Venezia Giulia',
    'VALLE D\'AOSTA': 'Valle d\'Aosta',
    'TRENTINO ALTO ADIGE': 'Trentino Alto Adige',
    'SICILIA': 'Sicilia',
    'CALABRIA': 'Calabria',
    'SARDEGNA': 'Sardegna',
    'MOLISE': 'Molise',
    'BASILICATA': 'Basilicata',
    'ITALIA': 'Italia'
}

# Rename and filter regions
aci_data['Regione'] = aci_data['Regione'].replace(rename_mapping)
filtered_aci_data = aci_data[aci_data['Regione'].isin(desired_regions)]

# Rename the column to "denominazione_regione"
filtered_aci_data.rename(columns={'Regione': 'denominazione_regione'}, inplace=True)
```
