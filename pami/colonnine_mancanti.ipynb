```python
# Check for duplicate chargePointIdentity values
duplicate_points = dfloc[dfloc.duplicated(subset=['chargePointIdentity'], keep=False)]

# Display the duplicate entries
print("Duplicate charge point identities:")
display(duplicate_points)
print('------------ \n')
print(len(duplicate_points))
```

```python
# Get the count of occurrences for each chargePointIdentity
duplicate_counts = dfloc['chargePointIdentity'].value_counts()

# Filter out only those with more than 1 occurrence (duplicates)
duplicates_only = duplicate_counts[duplicate_counts > 1]

# Display the duplicates with their counts
print("Duplicate charge point identities with their counts:")
print(duplicates_only)
```

```python
# Filter out incorrect entries: those not starting with 'IT' or NaN
incorrect_points = dfloc[~dfloc['chargePointIdentity'].astype(str).str.startswith('IT') | dfloc['chargePointIdentity'].isna()]

# Display the incorrect entries
print("Incorrect charge point identities:")
print(incorrect_points['chargePointIdentity'])
print()
```
